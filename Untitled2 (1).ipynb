```python
import numpy as np 
print("Dev Arora ")
from scipy import stats
grp1=[15,20,25,30] 
grp2=[10,15,20,25]
grp3=[5,10,15,20]
f_stats,p_value=stats.f_oneway(grp1,grp2,grp3)
print("F-statistic=",f_stats)
print("P-Value=",p_value) 
alpha=0.05
if p_value<alpha:
  print("Reject Null Hypothesis:There is a Significant difference between the means of the groups.")
else:
  print("Fail to Reject Null Hypothesis: There is no significant difference between the means of groups.")
```

```python
import math 
print("Dev Arora ")
def correlationcoefficient(X, Y, n):
    sum_X = 0
    sum_Y = 0
    sum_XY = 0
    squareSum_X = 0
    squareSum_Y = 0
    i = 0
    while i < n: 
        sum_X += X[i] 
        sum_Y += Y[i]
        sum_XY += X[i] * Y[i] 
        squareSum_X += X[i] * X[i] 
        squareSum_Y += Y[i] * Y[i]
        i += 1
        corr = (n * sum_XY - sum_X * sum_Y) / (math.sqrt((n * squareSum_X - sum_X) * (n * squareSum_Y - sum_Y * sum_Y)))
        return corr
X = [15, 18, 21, 24, 27]
Y = [25, 25, 27, 31, 32]
n = len(X)
print('{0:.6f}'.format(correlationcoefficient(X, Y, n))) 
```

```python
import math 
print("Dev Arora")
def mean(arr, n):
       sum = 0
       for i in range(0, n):
                sum = sum + arr[i] 
       return sum / n
def covariance(arr1, arr2, n): 
       sum = 0
       mean_arr1 = mean(arr1, n) 
       mean_arr2 = mean(arr2, n) 
       for i in range(0, n):
                sum = (sum + (arr1[i] - mean_arr1) * (arr2[i] - mean_arr2))
       return sum / (n - 1)
arr1 = [65.21, 64.75, 65.26, 65.76, 65.96]
n = len(arr1)
arr2 = [67.25, 66.39, 66.12, 65.70, 66.64]
m = len(arr2) 
if (m == n):
       print(covariance(arr1, arr2, m))
```
